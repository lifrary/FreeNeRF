```python
import pandas as pd

# Step 1: Read the values from each target text file.
def read_values_from_file(filename):
    with open(filename, 'r') as file:
        values = file.read().split()
    return [float(value) for value in values]

# List of numbers
numbers = [8, 21, 30, 31, 34, 38, 40, 41, 45, 55, 63, 82, 103, 110, 114]

# Base format string
#dtu = ['dtu3-end-0.9', 'dtu6-end-0.7', 'dtu9-end-0.2']
#dtu_number = ['43945', '87891', '131816'] 
dtu = 'dtu9-end-0.2'
dtu_number = '131836'

base_format = "out/freenerf/{}/{}/test_preds/{}.txt"

# Target metric names
metrics = [
    "metric_psnr_masked_"+dtu_number,
    "metric_ssim_masked_"+dtu_number,
    "metric_lpips_masked_"+dtu_number,
    "metric_avg_err_masked_"+dtu_number
]

# Data dictionary for the dataframe
data = {
    'psnr': [],
    'ssim': [],
    'lpips': [],
    'avg_arr': []
}

# Iterate through each number
for number in numbers:
    means = []
    for metric in metrics:
        filename = base_format.format(dtu, number, metric)
        values = read_values_from_file(filename)
        means.append(sum(values) / len(values))
    
    data['psnr'].append(means[0])
    data['ssim'].append(means[1])
    data['lpips'].append(means[2])
    data['avg_arr'].append(means[3])

# Step 3: Integrate the values into a dataframe.
df = pd.DataFrame(data, index=numbers)

# Calculate the mean of all rows and append it to the dataframe with row name 'mean'
mean_values = df.mean()
df.loc['mean'] = mean_values

# Step 4: Display the integrated dataframe.
print(df)

# If you want to save the dataframe to a new csv file:
df.to_csv(f"{dtu}.csv", index_label="Number")
```
